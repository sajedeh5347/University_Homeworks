<h2 style='direction:rtl; text-align: right'>
قسمت ۱.۳: پیاده سازی تابع norm
</h2>
$$\|x\| = \sqrt{\sum_{i=1}^{dim} x_i^2}$$
<p style='direction:rtl; text-align: right'>طوری تابع را پیاده سازی کنید که اندازه یک آرایه چند تایی از بردار ها را نیز حساب کند.</p>
<p style='direction:rtl; text-align: right'>به طور مثال اگر ورودی تابع ۲ بردار چند بعدی باشد خروجی یک آرایه ۲ تایی از اندازه بردار ها باشد.</p>

```python
import math
def vector_norm(x):
    ######## Write your code
    norm=[]
    for i in range(len(x)):
        sum=0
        for j in range(len(x.transpose())):
            sum=sum+(x[i,j])**2
        a=math.sqrt(sum)
        norm.append(a)
    
    ########
    return norm
```


<p style='direction:rtl; text-align: right'>در این قسمت تمام بردارها را به اندازه خود تقسیم کرده و بردارهای یکه حاصل را نمایش دهید.</p>

```python
######## Write your code
answer=[]
norm=vector_norm(k)
for i in range(len(k)):
    a1=[]
    for j in range(len(k.transpose())):
        a=k[i,j]/norm[i]
        a1.append(a)
    answer.append(a1)
    
########
print(len(answer))
print(answer)
```

<p style='direction:rtl; text-align: right'>
  صد بردار تصادفی ۱۰۰۰ بعدی از توزیع نرمال با میانگین صفر و انحراف معیار یک بسازید. سپس بردارها را به طولشان تقسیم کنید تا یکه شوند.

```python
# x is an array of dimension (100,1000)
n=100
dim=1000
######## Write your code 

answer= np.random.normal(0,1,size=(n,dim))
norm_answer=vector_norm(answer)
x=[]
for i in range(len(answer)):
    a2=[]
    for j in range(len(answer.transpose())):
        a=answer[i,j]/norm_answer[i]
        a2.append(a)
    x.append(a2)
    
########
print(len(x))
print(x)
########
```

<p style='direction:rtl; text-align: right'>
    با استفاده از دستور np.random.rand

</p>

```python
# y is an array of dimension (100,1000)
######## Write your code
x1=(np.random.rand(100))**0.001 
y=np.array(x)
for i in range(100):
    y[i,:]=y[i,:]*x1[i]
       

print(y)
print(y.shape)
########
```

<p style='direction:rtl; text-align: right'>
    میانگین طول بردارهای بدست آمده در قسمت قبل را پیدا کنید. چه نتیجه ای میگیرید؟
 نتیجه میگیریم که اکثر بردارها در نزدیکی پوسته قرار دارند.
</p>

```python
# mean1 is the average of norms of vectors in y
######## Write your code

norm=vector_norm(y)
norm=np.array(norm)
mean1=norm.mean()
print(mean1) 
########
```

<p style='direction:rtl; text-align: right'>
ضرب داخلی دو به دو بردارهای درون y را بدست آورید و سپس میانگین این اعداد را پیدا کنید. چه نتیجه‌ای میگیرید؟  
</p>

```python
# mean2 is the average of inner products of y[i] and y[j]
######## Write your code
s=0
num=0
for i in range(99):
    for j in range(i+1,100):
        k=0
        ab=0
        while k<1000:
            ab=ab+y[i,k]*y[j,k]
            k=k+1
        num=num+1
        s=s+ab
mean2=s/num
print(mean2)
    
########
```

$$Distance(X,Y) = \sqrt{\sum_{i=1}^{dim} ((x_i-y_i)^2)}$$

```python
def distance(X,Y):
    ######## Write your code
    
    dim=len(X.transpose())
    sum=0
    for i in range(dim):
        sum=sum+(X[i]-Y[i])**2
        
    dif =math.sqrt(sum) 
    
    ########
    return dif
```

قسمت ۴.۱: پیاده سازی تابع Project جهت تصویر کردن بردار
</h2>
<p style='direction:rtl; text-align: right'>این تابع ۲ بردار به عنوان ورودی میگیرد و اولی را بر روی دومی تصویر می کند.</p>
<p style='direction:rtl; text-align: right'>دقت شود ورودی اول می تواند آرایه ای از بردار ها باشد. در نتیجه کد شما باید همچنان کار کند.</p>

```python
def Project(X,Y):
    
    ######## Write
    projection=np.zeros((len(X),len(X.transpose())))
    for i in range(len(X)):
        sum=0
        N=0
        for j in range(len(X.transpose())):
            sum=sum+(X[i,j]*Y[0,j])
            N=N+Y[0,j]**2
        a=(sum/N)*Y
        projection[i,:]=a
    
    ########
    return projection
```

قسمت ۴.۲: پیاده سازی تابع Project_Subspace جهت تصویر کردن بردار روی زیرفضا
</h2>
<p style='direction:rtl; text-align: right'>
  این تابع یک بردار d بعدی و k بردار d بعدی را می‌گیرد و مختصات تصویر بردار اول را روی فضای تولید شده توسط k بردار بعدی محاسبه می‌کند. خروجی یک بردار k بعدی است.
</p>
<p style='direction:rtl; text-align: right'>دقت شود ورودی اول می تواند آرایه ای از بردار ها باشد. در نتیجه کد شما باید همچنان کار کند.</p>

```python
def Project_Subspace(X,W): # X is a vector of dimension d, W is an array of size (k,d)
    
    ######## Write your code
    
    projection =np.zeros((len(X),len(W)))
    for i in range(len(X)):
        Xp=np.zeros(len(W))
        for j in range(len(W)):
            sum=0
            for k in range(len(W.transpose())):
                sum=sum+X[i,k]*W[j,k]
            Xp[j]=sum
        projection[i,]=Xp
    
    ########
    return projection
```
